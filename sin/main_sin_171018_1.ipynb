```python
def display_multi_boxplots(start_feature_name, row, col):
    with plt.style.context(('seaborn-darkgrid')):
        f, axes = plt.subplots(row, col, figsize=(2.3*col, 3.5*row), sharey=True)

        nums = row*col
        start_idx = cols_cat.index(start_feature_name)

        for num, ax in zip(range(nums), axes.flat):
            try:
                sns.boxplot(x=df_x[cols_cat[start_idx + num]], y=df_y['Price'], palette="Set2", ax=ax)
            except:
                pass
```

# Real 변수의 leverage, outlier 탐색

```python
df_x_c_0 = sm.add_constant(df_x[cols_real[0]])
df_x_c_1 = sm.add_constant(df_x[cols_real[1]])
df_x_c_2 = sm.add_constant(df_x[cols_real[2]])
df_x_c_3 = sm.add_constant(df_x[cols_real[3]])
df_x_c_4 = sm.add_constant(df_x[cols_real[4]])
df_x_c_5 = sm.add_constant(df_x[cols_real[5]])
df_y_price = pd.DataFrame(df_y, columns=["Price"])
```

```python
model0 = sm.OLS(pd.DataFrame(df_y_price), pd.DataFrame(df_x_c_0))
result0 = model0.fit()

model1 = sm.OLS(pd.DataFrame(df_y_price), pd.DataFrame(df_x_c_1))
result1 = model1.fit()

model2 = sm.OLS(pd.DataFrame(df_y_price), pd.DataFrame(df_x_c_2))
result2 = model2.fit()

model3 = sm.OLS(pd.DataFrame(df_y_price), pd.DataFrame(df_x_c_3))
result3 = model3.fit()

model4 = sm.OLS(pd.DataFrame(df_y_price), pd.DataFrame(df_x_c_4))
result4 = model4.fit()

model5 = sm.OLS(pd.DataFrame(df_y_price), pd.DataFrame(df_x_c_5))
result5 = model5.fit()
```

```python
influence0 = result0.get_influence()
hat0 = influence0.hat_matrix_diag

influence1 = result1.get_influence()
hat1 = influence1.hat_matrix_diag

influence2 = result2.get_influence()
hat2 = influence2.hat_matrix_diag

influence3 = result3.get_influence()
hat3 = influence3.hat_matrix_diag

influence4 = result4.get_influence()
hat4 = influence4.hat_matrix_diag

influence5 = result5.get_influence()
hat5 = influence5.hat_matrix_diag
```

```python
plt.figure(figsize=(13, 10))
plt.subplot(3, 2, 1)
plt.stem(hat0)
plt.xlabel(cols_real[0])

plt.subplot(3, 2, 2)
plt.stem(hat1)
plt.xlabel(cols_real[1])

plt.subplot(3, 2, 3)
plt.stem(hat2)
plt.xlabel(cols_real[2])

plt.subplot(3, 2, 4)
plt.stem(hat3)
plt.xlabel(cols_real[3])

plt.subplot(3, 2, 5)
plt.stem(hat4)
plt.xlabel(cols_real[4])

plt.subplot(3, 2, 6)
plt.stem(hat5)
plt.xlabel(cols_real[5])

plt.tight_layout()
plt.show()
```
